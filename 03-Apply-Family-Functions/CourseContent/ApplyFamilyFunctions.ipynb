# Project Overview

You are working on a project for a metorology beureau. You have been supplied weather data for 4 cities in the US: Chicago, New York, Houston, and San Franciso.

You are required to deliver the following outputs:
    
* A table showing the annual averages of each observed metric for every city
* A table showing by how much temperature fluctuates each month from min to max in percent. Take min temp as the base.
* A table showing the annual maximums of each observed metric for each city
* A table showing the annual minimums of each observed metric for each city
* A table showing in which months the annual maximums of each metric were observed in every city.

In this section we will cover:

1. How the apply family of functions works
2. How to recreate the apply statement with a for loop
3. when to apply, lapply, or sapply
4. How to combine lapply with []
5. How to nest your own functions in apply type functions
6. How to nest apply type functions within each other
7. How to use the which.max and which.min functions

# Import the dataset

```r
Chicago = read.csv("Chicago-F.csv", row.names = 1)
NewYork = read.csv("NewYork-F.csv", row.names = 1)
Houston = read.csv("Houston-F.csv", row.names = 1)
SanFran = read.csv("SanFran-F.csv", row.names = 1)
```

## Let's convert these to matrices

```r
Chicago = as.matrix(Chicago)
NewYork = as.matrix(NewYork)
Houston = as.matrix(Houston)
SanFran = as.matrix(SanFran)
```

## Let's make a list

```r
Weather = list(Chicago = Chicago, NewYork = NewYork, Houston = Houston, SanFran = SanFran)
Weather
```

# Using apply()

```r
apply(Chicago, 1, mean)
```

```r
apply(Chicago, 1, max)
```

```r
apply(Chicago, 1, min)
```

# Recreating the apply function with loops

## Find the mean of every row

```r
output = NULL
for(i in 1:5){
    output[i]= mean(Chicago[i,])
}
names(output) = rownames(Chicago)
output
```

## Find the max of every row

```r
output = NULL
for(i in 1:5){
    output[i]= max(Chicago[i,])
}
names(output) = rownames(Chicago)
output
```

## Find the min of every row

```r
output = NULL
for(i in 1:5){
    output[i]= min(Chicago[i,])
}
names(output) = rownames(Chicago)
output
```

# Nesting Apply Functions

```r
lapply(Weather, apply, 1, max)
```

```r
lapply(Weather, function(x) apply(x,1,max))
```

# which.max() and which.min()

```r
which.max(Chicago[1,])
```

```r
names(which.max(Chicago[1,]))
```

```r
apply(Chicago,1,function(x) names(which.max(x)))
```

```r
lapply(Weather, function(y) apply(y,1,function(x) names(which.max(x))))
```

Now we have the max months for each city!
